```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib
```

```python
#Importing dataset

df = pd.read_csv("MR data - Compiled Data Set.csv")
df.head(2)
```

```python
# Dropping non-integer columns

df = df.drop(columns = ["Title","Text"])
df.head(2)
```

```python
# Setting all null values to 0

df = df.fillna(0)
df.head(2)
```

```python
# Renaming columns

df = df.rename(columns = {
    'Murder Happened with reason:\n1:Property/Land Disputes,\n2:Family Dispute\n3:Petty Quarrels,\n4:Money Disputes, \n5:Personal Vendetta, \n6:Love Affairs, \n7:Casteism\n8: Unknown/other': 'Reason of Murder',
    'Number of male victims(adult)':'Male victims',
    'Number of female victims(adult)':'Female victims',
    'Kidnnaping:\nNumber of child victims':'Number of kidnapped child victims',
    'Number of male(adult) ':'Kidnapped Males',
    'Number of female(adult)':'Kidnapped Females',
    'Crime Against Women ( Combined ):\n1. Murder with Rape \n2. Dowry Deaths(Sec. 3048)\n3. Suicide(sec 305/306)\n4. Kidnapping(All)\n5. Acid Attack(Sec. 326A IPC)\n6. Cruelty by Husband/in-laws((Sec.498 A IPC)\n7. Rape only(Sec. 376 or 511 IPC)\n8. Assault on Women with Intent to Outrage her Modesty (Sec. 354 IPC)\n9. Cyber Crimes against Women\n10. Protection of Children from Sexual Offences Act': 'Crimes against women(1)',
    'Number of childs invoved':'Number of children invoved'
    })
df.head(2)
```

```python
# Updating reason of Murder as integers based on a dictionary

dict = {
    0 : 0, # 0 represents no murders happened
    'Property Disputes': 1,
    'Land Disputes': 1,
    'Family Dispute': 2,
    'Petty Quarrels': 3,
    'Money Disputes': 4,
    'Personal Vendetta': 5,
    'Love Affairs': 6,
    'Casteism': 7,
    'Unknown reasons': 8,
    'Other reasons': 8
}
df["Reason of Murder"] = df["Reason of Murder"].apply(lambda x : dict[x])
df.head(15)
```

```python
# Adding new columns for multiple reasons

df["Crimes against women(2)"]=0
df["Crimes against women(3)"]=0
df = df[['City ',
 'Reason of Murder',
 'Number of child victims',
 'Male victims',
 'Female victims',
 'Number of kidnapped child victims',
 'Kidnapped Males',
 'Kidnapped Females',
 'Crimes against women(1)',
 'Crimes against women(2)',
 'Crimes against women(3)',
 'Number of adult victims',
 'Number of children invoved']]
df.head(2)
```

```python
# Breaking multiple reasons of 'Crimes against women' to different columns

for i in range(0 , len(df)):
    if ',' in str(df.iloc[i,8]):
        cl = df.iloc[i,8].split(',')
        df.iloc[i,8]=int(cl[0])
        df.iloc[i,9]=int(cl[1])
        if len(cl) > 2 :
            df.iloc[i,10]=int(cl[2])
    elif int(df.iloc[i,8])>100 :
        df.iloc[i,9]=int(df.iloc[i,8])%100
        df.iloc[i,8]=int(df.iloc[i,8])//100

df.head(15)
```

```python
# Analyzing City-wise crimes 

city1 = 0
city2 = 0
city3 = 0
for i in range(0 , len(df)):
    if df.iat[i,0] == "Ghaziabad":
        city1 = city1 + 1
    elif df.iat[i,0] == "Kanpur":
        city2 = city2 + 1
    elif df.iat[i,0] == "Lucknow":
        city3 = city3 + 1
city = list(("Ghaziabad","Kanpur","Lucknow"))     
count = list((city1,city2,city3))
plt.bar(city, count, color =['red','orange','blue'],width = 0.5)
plt.xlabel("CITY")
plt.ylabel("TOTAL CRIMES")
plt.title("CITY-WISE CRIMES")
plt.show()

```

```python
# Analyzing Reason-wise crimes 

# Dictionary for Reason for Murder
#1 'Property Disputes' or 'Land Disputes'
#2 'Family Dispute'
#3 'Petty Quarrels'
#4 'Money Disputes'
#5 'Personal Vendetta'
#6 'Love Affairs'
#7 'Casteism'
#8 'Unknown reasons' or 'Other reasons'

df["Reason of Murder"].value_counts()
graph0 = df["Reason of Murder"].value_counts().to_dict()
graph0.pop(0)

val = list(graph0.values())
key = list(graph0.keys())
plt.figure(figsize = (6,4))
plt.bar(range(len(graph0)),val,tick_label = key,width = 0.8,color='green')
plt.xlabel("REASON OF MURDER")
plt.ylabel("COUNT OF CRIME")
plt.title("REASON-WISE MURDERS")
plt.show()
```

```python
# Analyzing Reason-wise crimes against Women

# Dictionary for Reason of crime against Women
# 1. Murder with Rape 
# 2. Dowry Deaths
# 3. Suicide
# 4. Kidnapping
# 5. Acid Attack
# 6. Cruelty by Husband/in-laws
# 7. Rape only
# 8. Assault on Women with Intent to Outrage her Modesty 
# 9. Cyber Crimes against Women
# 10. Protection of Children from Sexual Offences Act
# 11. Others

for i in range(0 , len(df)):
    for j in range(0 , 11):
        if df.iat[i,8] == str(j):
            df.iat[i,8] = int(j)
    for j in range(0 , 11):
        if df.iat[i,9] == str(j):
            df.iat[i,9] = int(j)
    for j in range(0 , 11):
        if df.iat[i,10] == str(j):
            df.iat[i,10] = int(j)

df["Crimes against women(1)"].value_counts()

graph1 = df["Crimes against women(1)"]
graph2 = df["Crimes against women(2)"]
graph3 = df["Crimes against women(3)"]
graph4 = pd.concat([graph1,graph2,graph3])
graph5 = graph4.value_counts().to_dict()
graph5.pop(0)

val = list(graph5.values())
key = list(graph5.keys())
plt.figure(figsize = (8,6))
plt.bar(range(len(graph5)),val,tick_label = key,width = 0.8,color='purple')
plt.xlabel("REASON OF CRIME AGAINST WOMEN")
plt.ylabel("COUNT OF CRIME")
plt.title("REASON-WISE CRIMES AGAINST WOMEN")
plt.show()
```

```python
list1 = []
for i in range(0 , len(df)):
    if df.iat[i,9] != 0 and df.iat[i,10] == 0:
        a = int(df.iat[i,8]) 
        b = int(df.iat[i,9])
        if a < b :
            list1.append(f"{a},{b}")
        elif b < a :
            list1.append(f"{b},{a}")
list1.sort()
#print(list1)

count = 0
i = 0
pairs = []
number = []
z = True
while z:
    y = True
    while y:
        if list1[i] == list1[i + 1]:
            count = count + 1
            i = i + 1
            y = True
        else:
            y = False
    pairs.append(f"{list1[i]}")
    number.append(f"{count+1}")
    i = i + 1
    count = 0
    if i > (len(list1)-2):
        z = False
for i in range(0,len(number)):
    c = int(number[i])
    number[i] = c
print(pairs)
print(number)
plt.figure(figsize = (10,6))
plt.bar(pairs,number,width = 0.8,color='purple')
plt.xlabel("PAIRS OF CRIMES AGAINST WOMEN")
plt.ylabel("COUNT OF CRIME")
plt.title("MULTIPLE CRIMES AGAINST WOMEN")
plt.show()
```
